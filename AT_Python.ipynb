**Exercício 1**

Defina uma função chamada square_range com 3 parâmetros (os mesmos da função range):

start: nº inicial da sequência gerada pela função range (c/ valor padrão 0)
stop: nº onde a sequência gerada pela função range para (não incluído na sequência)
step: diferença entre números consecutivos na sequência (c/ valor padrão 1)
A função square_range deve usar list comprehension para criar uma lista com os quadrados dos números gerados pela função range (com os argumentos recebidos) e retorná-la.

Documente a função square_range usando uma docstring.

Use a função square_range para criar 4 listas de acordo com as especificações abaixo e as imprima:

Definindo valores para os 3 parâmetros
Definindo apenas o valor para o parâmetro stop
Definindo valores apenas para os parâmetros start e stop
Definindo valores apenas para os parâmetros stop e step

```python
def square_range(start=0, stop=None, step=1):
    """
    Retorna uma lista contendo os quadrados dos números gerados pela função range, com os argumentos fornecidos.

    :param start: Número inicial da sequência (padrão 0).
    :param stop: Número onde a sequência para (não incluído na sequência).
    :param step: Diferença entre números consecutivos na sequência (padrão 1).
    :return: Uma lista com os quadrados dos números gerados pela função range.
    """
    if stop is None:
        start, stop = 0, start

    return [x ** 2 for x in range(start, stop, step)]

# Definindo valores para os 3 parâmetros
result1 = square_range(2, 7, 2)

# Definindo apenas o valor para o parâmetro stop
result2 = square_range(5)

# Definindo valores apenas para os parâmetros start e stop
result3 = square_range(3, 10)

# Definindo valores apenas para os parâmetros stop e step
result4 = square_range(stop=6, step=2)

# Imprimir os resultados
print("Resultado 1:", result1)
print("Resultado 2:", result2)
print("Resultado 3:", result3)
print("Resultado 4:", result4)
```

Um professor de matemática resolveu desafiar seus alunos propondo uma representação diferente para operações aritméticas.

Na representação proposta, cada operação é representada por uma sequência de caracteres sem espaços, organizada da seguinte forma:

O 1º caractere representa o operador (+, -, * ou /)
Os caracteres com índices ímpares (a partir de 1) representam os dígitos do 1º operando
Os caracteres com índices pares (a partir de 2) representam os dígitos do 2º operando em ordem reversa
Por exemplo, na string '+846216', temos uma operação de soma, o 1º operando é 861 (+846216) e o 2º operando é 624 (+846216). Portanto, essa string representa a seguinte operação:

861 + 624

Defina uma função que receba uma operação representada conforme especificado acima e retorne seu resultado.

Teste sua função realizando algumas operações.

```python
def calcular_operacao(operacao):
    """
    Calcula uma operação aritmética representada por uma sequência de caracteres.

    :param operacao: Uma sequência de caracteres representando a operação.
    :return: Uma tupla contendo o operando1, o operador e o operando2, bem como o resultado da operação.
    """
    operador = operacao[0]
    operando1 = int(operacao[1::2])  # Dígitos ímpares representam o 1º operando
    operando2 = int(operacao[2::2][::-1])  # Dígitos pares (em ordem reversa) representam o 2º operando

    if operador == '+':
        resultado = operando1 + operando2
    elif operador == '-':
        resultado = operando1 - operando2
    elif operador == '*':
        resultado = operando1 * operando2
    elif operador == '/':
        if operando2 == 0:
            return "Erro: Divisão por zero."
        resultado = operando1 / operando2
    else:
        return "Operador inválido."

    return operando1, operador, operando2, resultado

# Teste de operações
operacao1 = '+846216'
operacao2 = '-753421'
operacao3 = '*239487'
operacao4 = '/968721'

resultado1 = calcular_operacao(operacao1)
resultado2 = calcular_operacao(operacao2)
resultado3 = calcular_operacao(operacao3)
resultado4 = calcular_operacao(operacao4)

print(f"{operacao1} = {resultado1[0]} {resultado1[1]} {resultado1[2]} = {resultado1[3]}")
print(f"{operacao2} = {resultado2[0]} {resultado2[1]} {resultado2[2]} = {resultado2[3]}")
print(f"{operacao3} = {resultado3[0]} {resultado3[1]} {resultado3[2]} = {resultado3[3]}")
print(f"{operacao4} = {resultado4[0]} {resultado4[1]} {resultado4[2]} = {resultado4[3]}")
```

Considere que a string recebida não contém espaços e que apenas um tipo de separador será usdo em cada string (não haverá alternância de separador em uma mesma string).

Defina outra função que receba uma lista de destinatários e os imprima em minúsculas como uma lista numerada (usando a função enumerate).

Teste suas funções com alguns exemplos.

```python
def separar_destinatarios(string_destinatarios, separador):
    """
    Esta função recebe uma string representando uma lista de destinatários separados por um separador específico
    e retorna uma lista de destinatários.

    :param string_destinatarios: A string com os destinatários.
    :param separador: O separador usado na string.
    :return: Uma lista de destinatários.
    """
    destinatarios = string_destinatarios.split(separador)
    return destinatarios

def imprimir_destinatarios(destinatarios):
    """
    Esta função recebe uma lista de destinatários e os imprime em minúsculas como uma lista numerada.

    :param destinatarios: A lista de destinatários.
    """
    for i, destinatario in enumerate(destinatarios, start=1):
        print(f"{i}. {destinatario.lower()}")

# Exemplo de uso
string_destinatarios = "DESTINATARIO1@gmail.com;destinatario2@gmail.com;destinatario3@gmail.com"
separador = ";"

lista_destinatarios = separar_destinatarios(string_destinatarios, separador)
imprimir_destinatarios(lista_destinatarios)
```

**Exercício 4**

O arquivo conjuntos.txt contém 10 conjuntos de números inteiros, cada um em uma linha, com os elementos separados por espaços em branco.

Escreva um programa que leia os dados do arquivo, calcule e imprima:

o resultado da interseção entre todos os conjuntos
o nº de elementos nessa interseção

```python
# Função para calcular a interseção entre todos os conjuntos em um arquivo
def calcular_intersecao_total_com_exibicao(arquivo):
    with open(arquivo, 'r') as file:
        # Lê todos os conjuntos do arquivo
        conjuntos = [set(map(int, linha.strip().split())) for linha in file]

    if not conjuntos:
        return set()  # Retorna um conjunto vazio se o arquivo estiver vazio

    # Inicializa a interseção total com o primeiro conjunto
    intersecao_total = conjuntos[0]
    print("Conjunto sendo verificado:", intersecao_total)

    # Calcula a interseção gradualmente e exibe o conjunto sendo verificado
    for i, conjunto in enumerate(conjuntos[1:], start=2):
        print("Conjunto sendo verificado (passo", i, "):", conjunto)
        intersecao_total = intersecao_total.intersection(conjunto)

    return intersecao_total

# Nome do arquivo contendo os conjuntos
arquivo = 'conjuntos.txt'

# Calcula a interseção entre todos os conjuntos no arquivo com exibição
intersecao_total = calcular_intersecao_total_com_exibicao(arquivo)

# Imprime a interseção e o número de elementos na interseção
print("Interseção entre todos os conjuntos:", intersecao_total)
print("Número de elementos na interseção:", len(intersecao_total))
```

**Exercício 5**

O arquivo clientes.json contém os cadastros de clientes de uma empresa.

Escreva um programa que leia os dados do arquivo, calcule e imprima as idades mínima, máxima e média (com uma casa decimal) dos clientes cadastrados.

```python
import json

# Função para calcular as idades mínima, máxima e média
def calcular_idades_min_max_media(clientes):
    if not clientes:
        return None, None, None

    idades = [cliente['idade'] for cliente in clientes if 'idade' in cliente]
    if not idades:
        return None, None, None

    idade_minima = min(idades)
    idade_maxima = max(idades)
    idade_media = sum(idades) / len(idades)

    return idade_minima, idade_maxima, idade_media

# Nome do arquivo contendo os dados dos clientes (JSON)
arquivo_json = 'clientes.json'

try:
    # Lê os dados do arquivo JSON
    with open(arquivo_json, 'r') as file:
        dados = json.load(file)
        clientes = dados.get('clientes', [])

    # Calcula as idades mínima, máxima e média
    idade_minima, idade_maxima, idade_media = calcular_idades_min_max_media(clientes)

    if idade_minima is not None:
        print(f"Idade mínima: {idade_minima} anos")
        print(f"Idade máxima: {idade_maxima} anos")
        print(f"Idade média: {idade_media:.1f} anos")
    else:
        print("Não há dados de clientes no arquivo ou não há idades registradas.")

except FileNotFoundError:
    print(f"O arquivo '{arquivo_json}' não foi encontrado.")
except json.JSONDecodeError:
    print(f"Erro ao decodificar o arquivo JSON.")
```

**Exercício 7**

Altere a primeira função definida no Exercício 3 para que os e-mails na lista retornada estejam em minúsculas.

Use essa função para separar os e-mails dos clientes nos cadastros do arquivo clientes.json e gere um novo arquivo chamado novo_clientes.json onde a chave "email" de cada cliente seja substituída por uma chave "emails" cujo valor será a lista com os e-mails separados (e em minúsculas).

```python
import json

# Função para separar e-mails em minúsculas
def separar_emails(emails_str):
    emails = [email.strip().lower() for email in emails_str.replace(';', ',').split(',')]
    return emails

# Nome do arquivo de entrada contendo os dados dos clientes (JSON)
arquivo_entrada = 'clientes_q7.json'

try:
    # Lê os dados do arquivo JSON de entrada
    with open(arquivo_entrada, 'r') as file:
        dados = json.load(file)

    # Verifica se há dados de clientes no arquivo
    clientes = dados.get('clientes', [])
    if not clientes:
        print("Não há dados de clientes no arquivo.")
    else:
        # Processa cada cliente para separar os e-mails em minúsculas
        for cliente in clientes:
            if 'email' in cliente:
                cliente['emails'] = separar_emails(cliente['email'])
                del cliente['email']

        # Cria um novo arquivo JSON com os dados modificados
        novo_arquivo = 'novo_clientes.json'
        with open(novo_arquivo, 'w') as novo_file:
            json.dump(dados, novo_file, indent=4)

        print(f"Novo arquivo '{novo_arquivo}' gerado com sucesso.")

except FileNotFoundError:
    print(f"O arquivo '{arquivo_entrada}' não foi encontrado.")
except json.JSONDecodeError:
    print(f"Erro ao decodificar o arquivo JSON.")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```

O gênero do cliente pode ter um dos seguintes valores no arquivo JSON:

"M" (Masculino)
"F" (Feminino)
null (Não-binário/Não quis informar)
Adapte o programa desenvolvido no Exercício 5 para calcular e imprimir as idades mínima, máxima e média dos clientes por gênero (para cada uma das 3 categorias acima).

```python
import json

# Carregar o arquivo JSON com os clientes
with open('clientes_q9.json', 'r') as arquivo_json:
    dados = json.load(arquivo_json)

clientes = dados['clientes']

# Dicionários para armazenar as idades dos clientes por gênero
idades_masculino = []
idades_feminino = []
idades_nao_binario = []

# Percorrer a lista de clientes e agrupar as idades por gênero
for cliente in clientes:
    idade = cliente.get("idade")
    genero = cliente.get("genero")

    if genero == "M":
        idades_masculino.append(idade)
    elif genero == "F":
        idades_feminino.append(idade)
    elif genero is None:
        idades_nao_binario.append(idade)

# Função para calcular a média das idades
def calcular_media(idades):
    if len(idades) == 0:
        return None
    return sum(idades) / len(idades)

# Calcular estatísticas por gênero
min_idade_masculino = min(idades_masculino, default=None)
max_idade_masculino = max(idades_masculino, default=None)
media_idade_masculino = calcular_media(idades_masculino)

min_idade_feminino = min(idades_feminino, default=None)
max_idade_feminino = max(idades_feminino, default=None)
media_idade_feminino = calcular_media(idades_feminino)

min_idade_nao_binario = min(idades_nao_binario, default=None)
max_idade_nao_binario = max(idades_nao_binario, default=None)
media_idade_nao_binario = calcular_media(idades_nao_binario)

# Imprimir as estatísticas por gênero
print("Estatísticas por gênero:")
print("Masculino:")
print("Idade mínima:", min_idade_masculino)
print("Idade máxima:", max_idade_masculino)
print("Idade média:", media_idade_masculino)
print()
print("Feminino:")
print("Idade mínima:", min_idade_feminino)
print("Idade máxima:", max_idade_feminino)
print("Idade média:", media_idade_feminino)
print()
print("Não-binário/Não quis informar:")
print("Idade mínima:", min_idade_nao_binario)
print("Idade máxima:", max_idade_nao_binario)
print("Idade média:", media_idade_nao_binario)
```

```python
import json
from datetime import date

# A variável "hoje" contém a data de hoje.
# Para o dia do mês, use:  hoje.day
# Para o mês, use:         hoje.month
# Para o ano, use:         hoje.year
hoje = date.today()

# Nome do arquivo JSON para armazenar os cadastros
arquivo_json = "cadastros.json"

# Função para carregar os cadastros a partir do arquivo JSON
def carregar_cadastros():
    try:
        with open(arquivo_json, "r") as arquivo:
            return json.load(arquivo)
    except FileNotFoundError:
        return []

# Função para salvar os cadastros no arquivo JSON
def salvar_cadastros(cadastros):
    with open(arquivo_json, "w") as arquivo:
        json.dump(cadastros, arquivo, indent=4)

# Carrega os cadastros do arquivo JSON ao iniciar o programa
cadastros = carregar_cadastros()

def menu():
  # Apresenta o menu
  print()
  print('Digite o nº da opção desejada:')
  print('[1] Inserir novo cadastro')
  print('[2] Alterar CPF')
  print('[3] Trocar sobrenomes')
  print('[4] Remover cadastro')
  print('[5] Listar todos os cadastros')
  print('[6] Encerrar')

  # Lê a opção do usuário
  return input()

def id_valido(uid, min, max):
  return min <= int(uid) <= max

def nome_valido(nome):
  return len(nome) > 0 and not ' ' in nome

def cpf_valido(cpf):
  return cpf.isdigit() and len(cpf) == 11

def insere_cadastro(cadastros):
  # Lê o ID
  uid = input('ID (opcional): ')

  # Valida o ID
  if uid != '' and not id_valido(uid, 1, len(cadastros) + 1):
    print('ID inválido')
    return False

  # Lê o primeiro nome
  nome = input('Primeiro nome: ').capitalize()

  # Valida o primeiro nome
  if not nome_valido(nome):
    print('Nome inválido')
    return False

  # Lê o sobrenome
  sobrenome = input('Sobrenome (apenas o último): ').capitalize()

  # Valida o primeiro nome
  if not nome_valido(sobrenome):
    print('Sobrenome inválido')
    return False

  # Adiciona o sobrenome ao nome
  nome += ' ' + sobrenome

  # Lê o CPF
  cpf = input('CPF (apenas dígitos): ')

  # Valida o CPF
  if not cpf_valido(cpf):
    print('CPF inválido')
    return False

  # Lê a data de nascimento
  data_nasc = input('Data de nascimento (dd/mm/aaaa): ')

  if uid == '':  # ID não preenchido
    # Insere cadastro no final
    cadastros.append({'nome': nome, 'cpf': cpf, 'nasc': data_nasc})
  else:  # ID preenchido
    # Insere cadastro na posição correspondente ao ID
    cadastros.insert(int(uid) - 1, {'nome': nome, 'cpf': cpf, 'nasc': data_nasc})

  print('Cadastro inserido com sucesso.')

  return True

def altera_cpf(cadastros):
  # Lê o ID
  uid = int(input('ID: '))

  # Valida o ID
  if not id_valido(uid, 1, len(cadastros)):
    print('ID inválido')
    return False

  # Lê o CPF
  cpf = input('Novo CPF: ')

  # Valida o CPF
  if not cpf_valido(cpf):
    print('CPF inválido')
    return False

  # Atualiza o CPF
  cadastros[uid - 1]['cpf'] = cpf

  print('CPF atualizado com sucesso.')

  return True

def troca_sobrenomes(cadastros):
  # Lê o 1º ID
  uid1 = int(input('ID 1: '))

  # Valida o 1º ID
  if not id_valido(uid1, 0, len(cadastros)):
    print('ID inválido')
    return False

  # Lê o 2º ID
  uid2 = int(input('ID 2: '))

  # Valida o 2º ID
  if not id_valido(uid2, 0, len(cadastros)):
    print('ID inválido')
    return False

  # Divide os nomes correspondentes em primeiro nome e sobrenome
  nome1 = cadastros[uid1 - 1]['nome'].split()
  nome2 = cadastros[uid2 - 1]['nome'].split()

  # Troca os sobrenomes, atualizando os cadastros
  cadastros[uid1 - 1]['nome'] = nome1[0] + ' ' + nome2[1]
  cadastros[uid2 - 1]['nome'] = nome2[0] + ' ' + nome1[1]

  print('Sobrenomes trocados com sucesso.')

  return True

def remove_cadastro(cadastros):
  # Lê o ID
  uid = int(input('ID: '))

  # Valida o ID
  if not id_valido(uid, 0, len(cadastros)):
    print('ID inválido')
    return False

  #  Remove os valores do cadastro correspondente
  cadastros.pop(uid - 1)

  print('Cadastro removido com sucesso.')

  return True

def calcula_idade(data_nascimento):
  # Lista com nº de dias de cada mês
  dias_no_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

  # Separa e converte os campos da data de nascimento
  dia_nasc = int(data_nascimento[:2])
  mes_nasc = int(data_nascimento[3:5])
  ano_nasc = int(data_nascimento[6:])

  # Calcula a idade da pessoa em dias
  # Inicializa a contagem de dias em zero
  dias = 0

  # Para cada ano desde o nascimento da pessoa...
  for ano in range(ano_nasc, hoje.year + 1):
    # Define o 1º mês a ser contado no ano
    mes_inicial = 1

    # Se for ano de nascimento, começa a contar do mês de nascimento
    if ano == ano_nasc:
      mes_inicial = mes_nasc

    # Define o último mês a ser contado no ano
    mes_final = 12

    # Se for ano em que estamos, conta até o mês em que estamos
    if ano == hoje.year:
      mes_final = hoje.month

    # Para cada mês a ser contado neste ano...
    for mes in range(mes_inicial, mes_final + 1):
      # Se forem o ano e o mês em que estamos, soma o dia de hoje
      if ano == hoje.year and mes == hoje.month:
        dias += hoje.day
      else:
        # Senão, soma o nº de dias do mês
        dias += dias_no_mes[mes - 1]

        # Se for fevereiro e o ano for bissexto, soma mais 1 dia
        if mes == 2 and ((ano % 4 == 0 and ano % 100 != 0) or ano % 400 == 0):
          dias += 1

      # Se forem o ano e o mês de nascimento, subtrai o dia de nascimento
      if ano == ano_nasc and mes == mes_nasc:
        dias -= dia_nasc

  return dias

def imprime_cadastro(uid, cadastro, idade):
  # Imprime os dados do cadastro
  print()
  print('ID:', uid)
  print('Nome:', cadastro['nome'])
  print('CPF:', cadastro['cpf'])
  print('Data de nascimento:', cadastro['nasc'])
  print('Idade em dias:', idade)

def lista_cadastros(cadastros):
  # Se não há nenhum cadastro, imprime mensagem informando
  if len(cadastros) == 0:
    print('Nenhum cadastro')
    return False

  # Para cada cadastro...
  for id, cadastro in enumerate(cadastros, 1):
    # Imprime os dados do cadastro
    idade = calcula_idade(cadastro['nasc'])
    imprime_cadastro(id, cadastro, idade)

  return True

# Listas para armazenar os dados dos cadastros
cadastros = []

while True:
  opcao = menu()
  if opcao == '1':
    insere_cadastro(cadastros)
  elif opcao == '2':
    altera_cpf(cadastros)
  elif opcao == '3':
    troca_sobrenomes(cadastros)
  elif opcao == '4':
    remove_cadastro(cadastros)
  elif opcao == '5':
    lista_cadastros(cadastros)
  elif opcao == '6':
    # Encerra o programa e salva os cadastros no arquivo JSON
    salvar_cadastros(cadastros)
    break
  else:
    print('Opção inválida!')
```

**Exercício 12**

À medida que seus barcos se aproximam da terra, os elfos começam a fazer um inventário de seus suprimentos. Uma consideração importante é a comida - em particular, o número de calorias que cada elfo carrega.

Os elfos se revezam anotando o número de calorias contidas nas várias refeições, lanches, rações, etc. que trouxeram consigo, um item por linha. Cada elfo separa seu próprio inventário do inventário do elfo anterior (se houver) por uma linha em branco.

Por exemplo, suponha que os elfos terminem de escrever as calorias de seus itens e terminem com a seguinte lista:

1000
2000
3000
4000
5000
6000
7000
8000
9000
10000

Esta lista representa as calorias da comida transportada por cinco elfos:

O 1º elfo está carregando comida com 1000, 2000, e 3000 calorias, um total de 6000 calorias.
O 2º elfo está carregando um alimento com 4000 calorias.
O 3º elfo está carregando comida com 5000 e 6000 calorias, um total de 11000 calorias.
O 4º elfo está carregando comida com 7000, 8000, e 9000 calorias, um total de 24000 calorias.
O 5º elfo está carregando um alimento com 10000 calorias.
Caso os elfos fiquem com fome e precisem de lanches extras, eles precisam saber a qual elfo perguntar: eles gostariam de saber quantas calorias estão sendo transportadas pelo elfo que carrega mais calorias. No exemplo acima, é 24000 (4º elfo).

Encontre o elfo carregando mais calorias. Quantas calorias totais ele está carregando?

A entrada para seu programa será um arquivo contendo uma lista no formato especificado acima. Teste seu programa com os arquivos exemplo1.txt e exemplo2.txt

```python
def encontrar_elfo_com_mais_calorias(nome_arquivo):
    try:
        with open(nome_arquivo, 'r') as arquivo:
            linhas = arquivo.readlines()

        calorias_elfo_atual = 0
        max_calorias_elfo = 0

        for linha in linhas:
            linha = linha.strip()  # Remove espaços em branco e quebras de linha

            if linha:  # Se a linha não estiver em branco
                calorias_elfo_atual += int(linha)
            else:  # Linha em branco separa os inventários dos elfos
                max_calorias_elfo = max(max_calorias_elfo, calorias_elfo_atual)
                calorias_elfo_atual = 0

        # Verifica novamente após o loop (para o último elfo)
        max_calorias_elfo = max(max_calorias_elfo, calorias_elfo_atual)

        return max_calorias_elfo

    except FileNotFoundError:
        print(f'O arquivo "{nome_arquivo}" não foi encontrado.')

# Teste com o arquivo exemplo1.txt
resultado = encontrar_elfo_com_mais_calorias('exemplo1.txt')
print(f'O elfo com mais calorias está carregando um total de {resultado} calorias.')

# Teste com o arquivo exemplo2.txt
resultado = encontrar_elfo_com_mais_calorias('exemplo2.txt')
print(f'O elfo com mais calorias está carregando um total de {resultado} calorias.')
```
